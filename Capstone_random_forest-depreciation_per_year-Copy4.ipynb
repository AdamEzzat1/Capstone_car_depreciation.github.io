# Data cleaning

```python
KBB_df = pd.read_csv('KBB_df(6).csv')
```

```python
del KBB_df['Unnamed: 0']
del KBB_df['Unnamed: 0.1']
del KBB_df['Year']
del KBB_df['Maker']
del KBB_df['Model']
```

```python
KBB_df['depreciation'] = KBB_df['depreciation'].astype(float)
```

```python
#valuecounts on safety rating then divide by len of the whole dataframe 
#as there is only three ratings for safety which are nothing, 4, and 5. Replace nothing with zero so the algorithm will work
```

```python
KBB_df['SafetyRating'] = KBB_df['SafetyRating'].astype(str)
KBB_df['SafetyRating'] = KBB_df['SafetyRating'].fillna('nothing')
```

```python
KBB_df['Price_Used_USD'] = KBB_df['price in USD'].astype(float)
KBB_df['Price_New_USD'] = KBB_df['New_Price'].astype(float)
KBB_df['brand'] = KBB_df['brand'].astype(str)
```

```python
KBB_df['depreciation'] = (KBB_df['Price_Used_USD']/KBB_df['Price_New_USD']).apply(lambda x: 1 - x)
```

```python
KBB_df['depreciation_per_year'] = KBB_df['depreciation']/KBB_df['years since new']
```

```python
KBB_df['expected_depreciation'] = KBB_df['years since new'].apply(lambda x: x*KBB_df['depreciation_per_year'].mean())
```

```python
KBB_df.head()
```

```python
KBB_df['MPG'] = KBB_df['MPG'].str.replace(r'MPG', '')
KBB_df['HP'] = KBB_df['HP'].str.replace(r'HP', '')
```

```python
KBB_df['MPG'] = KBB_df['MPG'].astype(float)
```

```python
KBB_df['HP'] = KBB_df['HP'].astype(float)
```

```python
KBB_df.number_of_features[KBB_df.number_of_features <= (4)] = 0
KBB_df.number_of_features[KBB_df.number_of_features > (4)] = 1

```

```python
from scipy import stats 
```

```python
grouped3 = KBB_df.groupby('number_of_features')
#n0 = grouped3.count().loc[0][0]
#n1 = grouped3.count().loc[1][0]

sample_A = KBB_df[KBB_df['number_of_features'] == 0]
sample_B = KBB_df[KBB_df['number_of_features'] == 1]

#passing list of the depreciation of everything with bluetooth and everything without bluetooth 

t_result = stats.ttest_ind(sample_A['depreciation_per_year'], sample_B['depreciation_per_year'])
print(t_result)
```

the pvalue is below .05 so we can reason that the null hypothesis is not statistically significant, 
The null hypothesis was that cars more than 4 features would deprecieate slower
